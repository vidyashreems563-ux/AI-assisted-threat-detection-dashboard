```python
%%sql

DROP TABLE IF EXISTS CyberThreats;

CREATE TABLE CyberThreats (
ID INT,
Threat_Type VARCHAR(50),
Risk_Score INT
);

INSERT INTO CyberThreats VALUES
(1,'Login Failure',20),
(2,'Malware',85),
(3,'Phishing',70),
(4,'Suspicious Login',60);
```

```python
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
```

```python
%sql SELECT * FROM CyberThreats
```

```python
%sql SELECT * FROM CyberThreats WHERE Risk_Score > 50
```

```python
%sql SELECT AVG(Risk_Score) AS Average_Risk FROM CyberThreats
```
