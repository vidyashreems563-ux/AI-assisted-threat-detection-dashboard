```python


# Import library
import pandas as pd

# Create sample threat data
data = {
    "Threat": ["Login Failure", "Malware", "Phishing", "Suspicious Login"],
    "Risk_Score": [20, 85, 70, 60]
}

# Convert into DataFrame
df = pd.DataFrame(data)

# Display data
print("Threat Data:")
print(df)

# Calculate average risk
avg = df["Risk_Score"].mean()

print("\nAverage Risk Score:", avg)

# Filter high-risk threats
high_risk = df[df["Risk_Score"] > 50]

print("\nHigh Risk Threats:")
print(high_risk)
```
